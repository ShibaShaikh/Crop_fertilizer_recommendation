```python
features = crop_df.columns.to_list()
features.remove('label')
print(features)
```

```python
# visualize the distribution of each feature
fig, ax = plt.subplots(3, 3, figsize = (15,15))  # 3,3 = row,column
for i, subplot in zip(features, ax.flatten()):
    sns.histplot(crop_df[i], ax= subplot, kde= True)
    subplot.set_title(i)  # i column names

plt.tight_layout()
plt.show() 
```

```python
# plotting scatter plot
fig, ax = plt.subplots(3, 3, figsize = (15,15))
for i, subplot in zip(features, ax.flatten()):
    sns.histplot(x= i, y= 'label', data= crop_df, ax= subplot) 

plt.tight_layout()
plt.show() 
```

```python
# finding oyliers - boxplot
fig, ax = plt.subplots(3, 3, figsize = (15,15)) 
for i, subplot in zip(features, ax.flatten()): 
    sns.boxplot(crop_df[i], ax= subplot) 

plt.tight_layout()
plt.show() 
```

```python
# correlation of features
num_cols = crop_df.select_dtypes(include= [np.number])
num_cols
```

```python
corr_matrix = num_cols.corr()
corr_matrix
```

```python
# visualize with heatmap
sns.heatmap(corr_matrix)
```

```python
sns.heatmap(corr_matrix, annot = True)  # values inside color
```

```python
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.show()
```
